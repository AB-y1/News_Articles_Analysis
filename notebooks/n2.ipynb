first merge the news and rating dataset to a new one called combined_df

```python

# Define a list of CSV filenames 
csv_files = [r"C:\Users\HP\Downloads\week00_data\data.csv", r"C:\Users\HP\Downloads\week00_data\rating.csv"]

# Read all CSV files into a list of DataFrames
all_dataframes = []
for filename in csv_files:
  df = pd.read_csv(filename)
  all_dataframes.append(df)

# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(all_dataframes, ignore_index=True)

# Now 'combined_df' contains the data from all CSV files
print(combined_df)
```

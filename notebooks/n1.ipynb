```python

# Clean category column (assuming country names are all uppercase)
news['category'] = news['category'].str.upper()

# Clean title column by converting to uppercase and removing extra spaces
news['title_cleaned'] = news['title'].str.upper().str.strip()

# Extract words from the title column (consider splitting by delimiters like commas or hyphens if needed)
news['title_words'] = news['title_cleaned'].str.split()

# Helper function to check if category matches any word in the title
def category_in_title(category, title_words):
  category = row['category']  # Assuming category is a string column
  title_words = row['title_words']
  return category.upper() in title_words  # Perform case-insensitive check


# Apply the function to each row using vectorized apply
news['category_in_title'] = news.apply(lambda row: category_in_title(row['category'], row['title_words']), axis=1)

# Filter rows where category matches a word in the title
matched_titles = news[news['category_in_title'] == True]

# Print the DataFrame showing matching titles
print(matched_titles)
```
